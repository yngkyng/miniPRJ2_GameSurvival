> test1 데이터셋

```python
# 데이터 불러오기 & 전처리
import pandas as pd
label = pd.read_csv("c:/vscode/미니프로젝트/2_분류_프로젝트/data/test1_label.csv")

df = label
df = df.drop(['survival_time', 'amount_spent'], axis=1)
df = df.sort_values('acc_id')
df
```

```python
# 유저 일별 결제 금액
payment = pd.read_csv("c:/vscode/미니프로젝트/2_분류_프로젝트/data/test1_payment.csv")
# 캐릭터 간 일별 거래 이력
trade = pd.read_csv("c:/vscode/미니프로젝트/2_분류_프로젝트/data/test1_trade.csv")
trade1 = trade.drop(['time','server','source_char_id','target_char_id','item_type','item_amount', 'item_price'], axis = 1)
# 날짜, 유저 아이디 기준으로 거래횟수(사고/팔고) 구하기
trade_sell = trade1.groupby(['day','source_acc_id'])['type'].count().reset_index()
trade_buy = trade1.groupby(['day','target_acc_id'])['type'].count().reset_index()
trade_sell = trade_sell.rename(columns={'source_acc_id':'acc_id', 'type':'sell_cnt'})
trade_buy = trade_buy.rename(columns={'target_acc_id':'acc_id', 'type':'buy_cnt'})

# 캐릭터 일일 활동
activity = pd.read_csv("c:/vscode/미니프로젝트/2_분류_프로젝트/data/test1_activity.csv")
activity1 = activity.drop(['char_id', 'server'], axis = 1)

# 캐릭터 일일 전투 활동
combat = pd.read_csv("c:/vscode/미니프로젝트/2_분류_프로젝트/data/test1_combat.csv")
combat1 = combat.drop(['char_id', 'server', 'class'], axis = 1)

# 캐릭터 소속 혈맹 구성원들의 일일 전투 정보
pledge = pd.read_csv("c:/vscode/미니프로젝트/2_분류_프로젝트/data/test1_pledge.csv")
pledge1 = pledge.drop(['char_id', 'server', 'pledge_id'], axis = 1)
```

```python
df1 = payment.merge(trade_sell, on=['day', 'acc_id'], how='outer')
df1 = df1.merge(trade_buy, on=['day', 'acc_id'], how='outer')
df1 = df1.merge(activity1, on=['day', 'acc_id'], how='outer')
df1 = df1.merge(combat1, on=['day', 'acc_id'], how='outer')
df1 = df1.merge(pledge1, on=['day', 'acc_id'], how='outer')
df1 = df1.merge(df, on='acc_id', how='inner').sort_values(by=['day', 'acc_id'])
df1
```

```python
df1['acc_id'].nunique()
```

```python
# nan값 0으로 처리
df1.fillna(0, inplace=True)
df = df1
```

```python
# 변수 지정
train_cols = df.columns[2:36]
X = df[train_cols]
y = df['survive']

# 언더샘플링
from imblearn.under_sampling import RandomUnderSampler
X_sample, y_sample = RandomUnderSampler(random_state=0).fit_resample(X,y)
X_samp = pd.DataFrame(data=X_sample, columns=train_cols)
y_samp = pd.DataFrame(data=y_sample, columns=['survive'])
df_samp = pd.concat([X_samp, y_samp], axis=1)

# 변수 재지정
X=df_samp[train_cols]
y=df_samp["survive"]

# 학습용, 검증용 구분
# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(
#     X, y, test_size=0.2, stratify=y, random_state=0)

# 스케일링
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
# X_train = scaler.transform(X_train)
# X_test = scaler.transform(X_test)
```
